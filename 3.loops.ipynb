```python
#Given a number, reverse its digits (e.g., 3485 → 5843).
n = input("Enter a number to reverse: ")
rev = ""
for digit in n:
    rev = digit + rev
print(f"{n}={rev}")

#another way
num = int(input("Enter a number: "))

rev = 0
while num > 0:
    digit = num %10
    rev = rev *10+digit
    num //=10
    
print("Reverse",rev)

```

```python
#Check if a number is palindrome (121 → yes, 123 → no).
num = int(input("Enter a number to check if it is palindrome or not: "))
temp = num
rev = 0

while temp >0:
    digit = temp %10
    rev = rev * 10 + digit
    temp //=10

print(rev,num)
if rev == num:
    print("It's a Palindrome")
else:
    print("It's not a Palindrome")
```

Pattern

```python
#rectangle
for number in range(5):
    for j in range(5):
        print("* ",end="")
    print()
```

```python
#Right Angle:

num = int(input("Enter the number of rows: "))
for i in range(num):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
```

```python
#decreasing triangle
for i in range(5):
    for number in range(i,5):
        print("* ",end='')
    print()
```
