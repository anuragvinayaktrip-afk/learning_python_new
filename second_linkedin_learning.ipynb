```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if year % 400 == 0:
        leap = True
    elif year % 100 == 0:
        leap = False
    elif year % 4 == 0:
        leap = True
    else:
        leap = False
    return leap
        
year = int(input())
print(is_leap(year))
```

```python
n = int(input("Enter a number: "))
result = 0
for i in range(1, n+1):
    digits = len(str(i))         # number of digits in i
    result = result * (10**digits) + i

print(result)
    
```
